```python
import numpy as np
```

```python
X = np.random.rand(100,21)
X_transpose = X.transpose()
y = np.random.rand(100)
```

```python
print(X)
print(X.shape)
```

```python
print(X_transpose)
print(X_transpose.shape)
```

```python
part1 = np.dot(X_transpose, X)
part1_inverse = np.linalg.inv(part1)

part2 = np.dot(part1_inverse, X_transpose)
W_min = np.dot(part2, y)

print(W_min.shape)
print(W_min)
```
